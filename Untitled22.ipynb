```python
import numpy as np
```

```python
karachi=np.random.randint(low=20,high=100,size=[20,])
name=np.random.choice(['wasif','murtaza','haziq','ahsan'],20)
random=np.random.choice([10,11,13,14,14],20)
```

```python
a=list(zip(karachi,name,random))
```

```python
import pandas as pd
```

```python
dp=pd.DataFrame(data=a,columns=['karachi','name','random'])
```
